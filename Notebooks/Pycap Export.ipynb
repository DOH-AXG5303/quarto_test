# Gathered Survey Data

## Records

Records are the set of information for a unique participant and are assigned a unique record ID. Each record is composed of a number of fields (pieces of data), which can be spread across multiple forms (instruments) per record.  

Exporting records returns all of the data entered into the record(s) of interest. By default, all records in the project will be returned. A subset of records can be specified.  

The labels for the questions can be exported rather than the variable names. Similarly, the labels for the data entered can be exported instead of the corresponding raw numeric values that REDCap uses. 

```python
data = project.export_records(format_type='df') #all records with raw data values
data.head(10)
```

This exports all of the records ordered by record_id and includes one row for each event/repeating instrument. When the format type=df, there is a multi-index for both the record_id and redcap_event_name. **This maybe should not be the behavior - user should choose how to set the index or have a single index for record_id.** Exporting as a csv or json creates the index from the row number and the record_id and record_event_name are regular columns of data.

```python
#exporting all records with labeled data - csv
from io import StringIO

data_csv = StringIO(project.export_records(raw_or_label='label',
                                           format_type='csv')) 
df_csv = pd.read_csv(data_csv, sep=',')
df_csv.head(10)
```

Of note, when putting the exported data into a dataframe, any raw values for coded data that are integer strings in REDCap become floats with a decimal appended in python. If this exported dataset were to be re-imported into REDCap, errors will arise for all of the fields with the now float-type data when REDCap is expecting an integer. 

### Subsets of records or fields can be specified on export, or filtering can be done in python after exporting all of the data from the project.

#### Filtering by record

```python
#in python - filter from all data 
record2 = df_csv[df_csv["record_id"]==2]
record2
```

```python
#exporting only record 2 with labeled data - pycap
project.export_records(records='2',
                       raw_or_label='label',
                       format_type='df') 
```

```python
#exporting only record 2 with labeled data - csv - pycap
from io import StringIO

data_csv = StringIO(project.export_records(records='2',
                       raw_or_label='label',
                       format_type='csv')) 
df_csv = pd.read_csv(data_csv, sep=',')
df_csv
```

```python
#exporting only record 2 with labeled data and headers - csv - pycap
from io import StringIO

data_csv = StringIO(project.export_records(records='2',
                       raw_or_label='label',
                       raw_or_label_headers='label', #col names - raw is the variable name and label is the question
                       format_type='csv')) #raw_or_label_headers only works for csv 
df_csv = pd.read_csv(data_csv, sep=',')
df_csv
```

The raw_or_label_headers parameter (which should display the labeled field question/prompt rather than the variable name if set to label) only works when the export type is set to csv.

Interestingly, the event names here exported as the labeled names rather than the unique names as seen in the other examples, even though the event_name parameter was not specified. Let's explore that more:

```python
project.export_records(records='2', 
                       # raw_or_label='label', #this parameter changes the event name labeling as well
                       event_name='unique', #'label' is #default 
                       format_type='df')
```

```python
project.export_records(records='2', 
                       raw_or_label='label', #this parameter changes the event name labeling as well
                       event_name='unique', #'label' is #default - but this is overriden by above parameter
                       format_type='df')
```

The raw_or_label parameter (which is described as changing the raw vs coded values of multiple choice fields on export) changes the event labeling, and appears to override what is set with the event_name parameter, as this does not appear to have any effect on the event name exported when raw_or_label='label'.

## Files - optional attachments to individual records

File uploads are a unique field type in REDCap that accepts a variety of file types, including images and other documents. Unlike all other export methods, exporting files only works for one record at a time. 

If the project has repeating events (i.e. a longitudinal project), the event name must be specified. If the file of interest is in a repeat instance, the instance number must also be specified.

Doing a simple export of a file returns a tuple of binary data containing the content of the file and the content-type dictionary. Additional post-processing steps can be taken in python to convert the returned binary tuple into the file you expect to see.

```python
export_file_image = project.export_file(record="1", 
                                        field="test_upload", 
                                        event="case_intake_arm_1")
print(export_file_image)
```

The following code can be used to save and display the file in the format we expect:

```python
export_file_image = project.export_file(record="1", 
                                        field="test_upload", 
                                        event="case_intake_arm_1")
with open("files/test_api_export_image.png","wb") as binary_file:
    binary_file.write(export_file_image[0])
```

```python
from IPython.display import Image
Image("files/test_api_export_image.png", width=400)
```

This can also be done for other file types, such as pdfs.

```python
export_file_pdf1 = project.export_file(record="2", 
                                       field="test_upload", 
                                       event="case_intake_arm_1")
```

```python
export_file_pdf1 = project.export_file(record="2", 
                                       field="test_upload", 
                                       event="case_intake_arm_1")
with open("files/test_api_export_pdf1.pdf","wb") as binary_file: #the first argument in open() is the file path and file name for saving it
    binary_file.write(export_file_pdf1[0]) #this writes the binary information from the tuple to the file
```

```python
from IPython.display import IFrame
IFrame("files/test_api_export_pdf1.pdf", width=600, height=300)
```

If the file upload field to be exported is in a repeating instance, the instance number must be specified.

```python
export_file_pdf2 = project.export_file(record="2", 
                                       field="test_upload", 
                                       event="case_intake_arm_1", 
                                       repeat_instance=2)
with open("files/test_api_export_pdf2.pdf","wb") as binary_file: #the first argument in open() is the file path and file name for saving it
    binary_file.write(export_file_pdf2[0]) #this writes the binary information from the tuple to the file
```

## Instrument Event Map

Events are used in longitudinal data collection project designs. An event can be scheduled or unscheduled, be repeating or non-repeating, and may have one or more instrument(s) assigned to it. If an event repeats, all of the instruments in that event will repeat together.  

Longitudinal projects may also have arms (e.g., a control arm and a treatment arm for a clinical trial). The default number of arms for any project is one. For longitudinal projects, the arm name will be appended to the event name so that the events can be unique to each arm.  

This export method will return the arm number, event name with the arm appended, and the form (instrument) name.

```python
project.export_instrument_event_mappings(format_type='df')
```

This could be a way to filter the exported data if you are exporting by field/instrument. 

# Survey Settings and Other Information

## Users

Multiple people may be added as users to a project. Each user can be assigned certain privileges. These user rights include abilities such as instrument modifications, data entry or access, API access, and more.

```python
project.export_users(format_type='df') #exports project users (users' info, expiration, user privileges)
```

## User Roles

User roles are an optional feature that make it easier to assign privileges if there are several people who will need the same level of access, rather than assigning the same privileges manually. Roles are also a nice way to group and organize project users.

```python
project.export_user_roles(format_type='df') #exports the user roles of the project (role name, user privileges)
```

User role assignments can be exported to quickly see the role and DAG of each user.

```python
project.export_user_role_assignment(format_type='df')
```

## Data Access Groups (DAGs)

In addition to user roles for access control, DAGs control which records each user can access. Users assigned to a DAG can only access records assigned to that group, and are blinded to records outside of their group. DAGs are particularly useful in multi-site or multi-jurisdictional projects to restrict sites from viewing records from other sites.  

Users can be assigned to a DAG after being added to the project. Users can be in multiple DAGs. Users not assigned to any DAG have global access and can see all records in the project. 

```python
project.export_dags(format_type='df') #exports the names of the DAGs
```

```python
project.export_user_dag_assignment(format_type='df') #exports users and their assigned DAGs (username and DAG)
```

## Logging

Logging captures all changes made to the project, including management/design of instruments, data changes, API activity, and user changes. Exports of the logging history can be filtered in various ways (by user, record, dag, activity type, or time period).

```python
project.export_logging(format_type='df') #all logging history
```
