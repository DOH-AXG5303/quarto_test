```python
import redcap
import json
import requests
import pandas as pd
```

# Project Setup

```python
api_key = json.load(open('../PyCap/json_api_data.json'))
api_token = api_key['dev_token']['262']
api_url = api_key['dev_url']
```

```python
project = redcap.Project(api_url, api_token)
```

# Gathered Survey Data

## Records

```python
data = {
    'token': api_token,
    'content': 'record',
    'action': 'export',
    'format': 'json',
    'type': 'flat',
    'csvDelimiter': '',
    'records[0]': '2', #chooses which records to export (default if unspecified is all records)
    'rawOrLabel': 'label', #will display the number if 'raw', and the actual option selected if 'label' (for radio buttons/dropdowns) 
    'rawOrLabelHeaders': 'raw', #col names - raw is the variable name and label is the field's question
    'exportCheckboxLabel': 'true', #will just display "checked/unchecked" if false, and will display the actual option checked if true 
    'exportSurveyFields': 'false', #only relevant when surveys are enabled - exports additional fields like timestamps 
    'exportDataAccessGroups': 'false', #only relevant when DAGs are used in the project 
    'returnFormat': 'json'
}
r = requests.post('https://dev-redcap.doh.wa.gov/api/',data=data)
df_json = r.json()
df = pd.DataFrame.from_records(df_json)
df
```
